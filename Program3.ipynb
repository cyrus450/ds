```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Downloads/auto+mpg/auto-mpg.csv')
df.head()
```

```python


cats = list(df.select_dtypes(include=['object']).columns)
nums = list(df.select_dtypes(exclude=['object']).columns)

print("Categorical variables:", cats)
print("Numerical variables:", nums)
```

```python
for col in nums:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=20)
    plt.title(f"{col} Histogram")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['weight'], y=df['mpg'])
#sns.scatterplot(x='weight' , y='mpg' , data=df)
plt.title("Relationship: Weight vs MPG")
plt.xlabel("Weight (lbs)")
plt.ylabel("MPG")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='origin', data=df)
plt.title("Count by Origin")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.pointplot(x='origin', y='mpg', data=df)
plt.title("MPG Variation by Origin")
plt.show()
```
