```python
'''
Dictionaries are Python’s implementation of a data structure that is more generally known as an associative array. 
A dictionary consists of a collection of key-value pairs.
Each key-value pair maps the key to its associated value.

A dictionary is a collection which is ordered*, changeable and do not allow duplicates.
(As of Python version 3.7, dictionaries are ordered.
In Python 3.6 and earlier, dictionaries are unordered.)

A dictionary  can be defined by enclosing a comma-separated list of key-value pairs in curly braces ({}). A colon (:) separates each key from its associated value:

Syntax

d = {
    <key>: <value>,
    <key>: <value>,
      .
      .
      .
    <key>: <value>
}
'''
#Example to create dictionary
#Method-1

thisdict ={
  "Dept": "MCA",
  "Year_adm": "2022",
"Sem": ["first", "Second", "Third", "Fourth"]
}
print(thisdict)

#Method-2
'''
The dict() Constructor

It is also possible to use the dict() constructor to make a dictionary.
'''
thisdict = dict(name = "ABC", age = 36, country = "INDIA")
print(thisdict)

#Method -3
#Build dictionary incrementally
person = {}
type(person)
person['fname'] = 'Joe'
person['lname'] = 'Fonebone'
person['age'] = 51
person['spouse'] = 'Edna'
person['children'] = ['Ralph', 'Betty', 'Joey']
person['pets'] = {'dog': 'Fido', 'cat': 'Sox'}
print("\nthe dictionary created incrementally is",person)


#Duplicates are not allowed
#Dictionaries cannot have two items with the same key
#Duplicate values will overwrite existing values
thisdict ={
  "Dept": "MCA",
  "Year_adm": "2022",
  "Year_adm": "2023",
  "Sem": ["first", "Second", "Third", "Fourth"],
}
print(thisdict)

#Length of Dictionary
print(len(thisdict))
#Type of Dictionary
print(type(thisdict))

#access the items of a dictionary by referring to its key name, inside square brackets
thisdict ={
  "Dept": "MCA",
  "Year_adm": "2022",
  "Sem": ["first", "Second", "Third", "Fourth"]
}
x = thisdict["Sem"]
print("the element extracted from dictionary is",x)

'''
The numbers in square brackets appear as though they might be indices.
But they have nothing to do with the order of the items in the dictionary.
Python is interpreting them as dictionary keys.
If you define this same dictionary in reverse order,
you still get the same values using the same keys
'''
d = {0: 'a', 1: 'b', 2: 'c', 3: 'd'}
print('\nthe dictionary is',d)
print('d[0] and d[3] are',d[0],d[3])
d = {3: 'd', 2: 'c', 1: 'b', 0: 'a'}
print('the dictionary is',d)
print('d[0] and d[3] are',d[0],d[3])
#use of get()to access the elements of dictionary
x = thisdict.get("Dept")
print("\nthe element extracted from dictionary is",x)

#Get Keys : The keys() method will return a list of all the keys in the dictionary
x = thisdict.keys()
print("\nthe Keys extracted from dictionary is",x)

#Add a new item to the original dictionary
thisdict ={
  "Dept": "MCA",
  "Year_adm": "2022",
  "Sem": ["first", "Second", "Third", "Fourth"]
}
print("Before adding")
x = thisdict.keys()
print(x)
thisdict["Avg_pass_Res"] = "85"
print("After adding")
x = thisdict.keys()
print(x)
print(thisdict)

#Get Values: The values()method will return a list of all the values in the dictionary
x = thisdict.values()
print("the values of dictionary are",x)
thisdict ={
  "Dept": "MCA",
  "Year_adm": "2022",
  "Sem": ["first", "Second", "Third", "Fourth"]
}
print("Before change")
x = thisdict.values()
print(x)
thisdict["Year_adm"] = "2023"
print("After change")
x = thisdict.values()
print(x)

#Get Items:The items() method will return each item in a dictionary, as tuples in a list
x = thisdict.items()
print("\nitems in dictionary are",x)

#Check if Key Exists
thisdict ={
  "Dept": "MCA",
  "Year_adm": "2022",
  "Sem": ["first", "Second", "Third", "Fourth"]
}
if "Dept" in thisdict:
  print("\nYes, 'Dept' is one of the keys in the thisdict dictionary")
else:
  print("No, 'Dept' is not one of the keys in the thisdict dictionary")

#Update Dictionary

#The update() method will update the dictionary with the items from the given argument.

#The argument must be a dictionary, or an iterable object with key:value pairs
thisdict.update({"Dept": "Master of Computer Applications"})
print("\nupdated dictionary is", thisdict)

#Removing Items
#1. The pop() method removes the item with the specified key name
thisdict.pop("Year_adm")
print("\nafter pop()",thisdict)

#The popitem() method removes the last inserted item
thisdict.popitem()
print("\nafter popitem()",thisdict)

#The del keyword removes the item with the specified key name
thisdict ={
  "Dept": "MCA",
  "Year_adm": "2022",
  "Sem": ["first", "Second", "Third", "Fourth"]
}
del thisdict["Sem"]
print("\nafter deleting the key \"sem\"",thisdict)

#The clear() method empties the dictionary
thisdict.clear()
print("\nAfter clearing the dictionary",thisdict)

#deleting the entire dictionary
del thisdict
print("the dictionary is deleted")
#print(thisdict)



'''
Restrictions that dictionary keys must abide by:

A given key can appear in a dictionary only once
Duplicate keys are not allowed. A dictionary maps each key to a corresponding value,
so it doesn’t make sense to map a particular key more than once.
'''
```
